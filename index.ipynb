# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd 
import numpy as np
df = pd.DataFrame({"color":["blue","green","yellow","white"]})
df
```

```python
encoder = OneHotEncoder()

encoded = encoder.fit_transform(df).toarray()
```

```python
updated_encoded = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
updated_encoded = pd.concat([df,updated_encoded],axis=1)
updated_encoded
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
df1 = df.iloc[:,2:6]
df1.head()
```

```python
encoder1 = OneHotEncoder()
encoded = encoder1.fit_transform(df1).toarray()

tips_one_hot_encoding = pd.DataFrame(encoded,columns=encoder1.get_feature_names_out())
tips_one_hot_encoding
```

```python
pd.concat([df1,tips_one_hot_encoding],axis=1).tail(11)
```

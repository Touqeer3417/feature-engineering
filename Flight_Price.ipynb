### Featured Engineering Process start 

i can convnert date of jouneroy to date , month and year spearately so it can easily understand for machhine to learn

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# date month aand year are object data type i can convert into int data type 
df['Date']  = df['Date'].astype(int)
df['Month']  = df['Month'].astype(int)
df['Year']  = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["Arrival_Time"]= df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
df["Arrival_Time"]
```

```python
df["Arrival_Hour"] = df["Arrival_Time"].str.split(':').str[0]
df["Arrival_Minutes"] = df["Arrival_Time"].str.split(':').str[1]
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df['Arrival_Hour']  = df['Arrival_Hour'].astype(int)
df['Arrival_Minutes']  = df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["Dep_Time_Hour"] = df["Dep_Time"].str.split(":").str[0].astype(int)
df["Dep_Time_Minute"] = df["Dep_Time"].str.split(":").str[1].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python

def convert_to_minutes(time_str):
    hours = 0
    minutes = 0
    if 'h' in time_str:
        hours = int(time_str.split('h')[0].strip())
        time_str = time_str.split('h')[1]
    if 'm' in time_str:
        minutes = int(time_str.split('m')[0].strip())
    return hours * 60 + minutes

df['Duration_in_Minutes'] = df['Duration'].apply(convert_to_minutes)

df.head(1)
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["Total_Stops"].unique()

df["Total_Stops"]= df["Total_Stops"].map({'non-stop':0,'2 stops':2, '1 stop':1, '3 stops':3, np.nan:0, '4 stops':4 })
```
